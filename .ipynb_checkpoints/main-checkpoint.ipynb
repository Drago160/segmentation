```python
from torch.utils.data import DataLoader
import torch

from prepare_data.make_dataset import get_dataset

from modules.unet import Unet

from tqdm import tqdm
```

```python
from prepare_data.dataset import CVPPPDataset
from losses.inst_losses import LossKeeper
from train_func import train
```

```python
train_imgs, train_insts = get_dataset('train')
train_dataset = CVPPPDataset(train_imgs, train_insts)
train_dataloader = DataLoader(train_dataset, batch_size=1,
                        shuffle=True, num_workers=0)

val_imgs, val_insts = get_dataset('dev')
val_dataset = CVPPPDataset(val_imgs, val_insts)
val_dataloader = DataLoader(val_dataset, batch_size=1,
                        shuffle=True, num_workers=0)
```

```python
NUM_COLORS = 7
HALO_COEF = 0.02
HALO_MARGIN = 21

train_loss_keeper = LossKeeper(train_insts, 
                              NUM_COLORS,
                              HALO_COEF, 
                              HALO_MARGIN)

val_loss_keeper = LossKeeper(val_insts, 
                              NUM_COLORS,
                              HALO_COEF, 
                              HALO_MARGIN)
```

```python
model = Unet(3, NUM_COLORS)
opt = torch.optim.Adam(model.parameters(), lr = 1e-3)
```

```python
model, train_losses, val_losses = train(
    model, opt,
    train_dataloader, val_dataloader,
    train_loss_keeper, val_loss_keeper,
    5
)
```

```python
PATH = "model.pt"

torch.save({
            'model_state_dict': model.state_dict()
            }, PATH)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (7, 5))
ax.plot(train_losses)
ax.set(
    title = 'Train Loss',
    xlabel = 'iter'
)
plt.show()

fig, ax = plt.subplots(figsize = (7, 5))
ax.plot(val_losses)
ax.set(
    title = 'Val Loss',
    xlabel = 'iter'
)
```

```python
test_imgs, test_insts = get_dataset('test')
test_dataset = CVPPPDataset(test_imgs, test_insts)
test_dataloader = DataLoader(test_dataset, batch_size=1,
                        shuffle=True, num_workers=0)

test_loss_keeper = LossKeeper(test_insts, 
                              NUM_COLORS,
                              HALO_COEF, 
                              HALO_MARGIN)
```

```python
test_losses = []
model.eval()
for img_b, _, idx in test_dataloader:
    idx = int(idx)
    output = model(img_b)[0]
    loss = test_loss_keeper.forward(output, idx)
    test_losses.append(loss.item())
```

```python
fig, ax = plt.subplots(figsize = (7, 5))
ax.plot(val_losses)
ax.set(
    title = 'test Losses',
    xlabel = 'picture_num'
)
```
